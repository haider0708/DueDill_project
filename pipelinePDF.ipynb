```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time
import os
import requests

options = webdriver.ChromeOptions()
options.add_argument("--headless")  
# Set up the WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# Open the target webpage
url = "https://www.binance.com/en/research/analysis"
driver.get(url)
links = driver.find_elements(By.TAG_NAME, "a")

# Extract all article links
article_links = [link.get_attribute("href") for link in links]

# Folder to save PDFs
download_folder = "small test"
os.makedirs(download_folder, exist_ok=True)

pdf_count = 0  # Counter to track the number of PDFs downloaded

# Iterate through each article link
for article_url in article_links:
    if pdf_count >= 5:
        break  # Stop after downloading 5 PDFs

    driver.get(article_url)  # Open article page
    time.sleep(3)  # Wait for the page to load

    try:
        # Find the PDF link (Modify the selector if needed)
        pdf_link = driver.find_element(By.XPATH, "//a[contains(text(), 'DOWNLOAD THE FULL REPORT HERE')]")
        pdf_url = pdf_link.get_attribute("href")

        # Download the PDF file
        pdf_name = pdf_url.split("/")[-1]
        pdf_path = os.path.join(download_folder, pdf_name)

        print(f"Downloading: {pdf_url}")
        response = requests.get(pdf_url, stream=True)
        with open(pdf_path, "wb") as file:
            for chunk in response.iter_content(chunk_size=1024):
                file.write(chunk)

        print(f"Saved: {pdf_path}")
        pdf_count += 1  # Increment the counter

    except Exception as e:
        print(f"PDF not found on {article_url}: {e}")

# Close the browser
driver.quit()
print("First scraping complete :D !")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os
import requests

# Set up WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--headless")  
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# Set up download folder
download_folder = "small test"
if not os.path.exists(download_folder):
    os.makedirs(download_folder)

pdf_count = 0  # Counter to track the number of PDFs downloaded

for page_num in range(1, 121):
    url = f"https://messari.io/research/research-and-protocols?page={page_num}"
    driver.get(url)
    
    # Wait until the page content is loaded
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".link_root__Xj0Q3.MesCard-root.css-1g5o5jj"))
        )
        links = driver.find_elements(By.CSS_SELECTOR, ".link_root__Xj0Q3.MesCard-root")
        article_links = [link.get_attribute("href") for link in links if link.get_attribute("href")]

        print(f"Found {len(article_links)} articles on page {page_num}")

        for article_url in article_links:
            if pdf_count >= 10:
                break  # Stop after downloading 10 PDFs
            
            try:
                driver.get(article_url)
                time.sleep(1)  # Allow time for the page to load

                # Find all PDF links on the page
                pdf_links = driver.find_elements(By.XPATH, "//a[contains(@href, '.pdf')]")
                pdf_urls = [pdf.get_attribute("href") for pdf in pdf_links if pdf.get_attribute("href")]

                for pdf_url in pdf_urls:
                    if pdf_count >= 10:
                        break  # Stop after downloading 10 PDFs
                    
                    pdf_name = pdf_url.split("/")[-1]
                    pdf_path = os.path.join(download_folder, pdf_name)

                    if os.path.exists(pdf_path):
                        print(f"Skipping (already exists): {pdf_name}")
                        continue  # Skip download if file exists

                    print(f"Downloading: {pdf_url}")
                    response = requests.get(pdf_url, stream=True)
                    with open(pdf_path, "wb") as file:
                        for chunk in response.iter_content(chunk_size=1024):
                            file.write(chunk)

                    print(f"Saved: {pdf_path}")
                    pdf_count += 1  # Increment the counter

            except Exception as e:
                print(f"Error on {article_url}: {e}")

    except Exception as e:
        print(f"Error loading page {page_num}: {e}")

    if pdf_count >= 10:
        break  # Stop after downloading 10 PDFs

# Close the WebDriver
driver.quit()
print("third scraping complete :D !")
```

```python
import os
import PyPDF2

# Folder containing PDFs
download_folder = "small test"

# Function to check if a PDF is readable
def is_pdf_readable(pdf_path):
    try:
        with open(pdf_path, "rb") as file:
            reader = PyPDF2.PdfReader(file)
            if reader.pages:
                return True
    except Exception as e:
        print(f"Unreadable PDF detected: {pdf_path} - {e}")
    return False

# Iterate through PDF files and delete unreadable ones
for pdf_file in os.listdir(download_folder):
    if pdf_file.lower().endswith(".pdf"):
        pdf_path = os.path.join(download_folder, pdf_file)
        if not is_pdf_readable(pdf_path):
            os.remove(pdf_path)
            print(f"Deleted unreadable PDF: {pdf_file}")

print("Unreadable PDF cleanup complete.")
```

```python
import os
import pandas as pd
from pdfminer.high_level import extract_text
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import joblib
import hashlib
import shutil

CATEGORIES = {
    "Whitepaper": ["whitepaper", "protocol", "consensus", "blockchain"],
    "Research": ["research", "study", "academic", "analysis"],
    "Regulatory": ["compliance", "regulation", "legal", "SEC", "AML", "KYC"],
    "Project Documentation": ["API", "developer", "documentation", "smart contract"],
    "Exchange Reports": ["trading", "market", "liquidity", "exchange"],
    "ICO Documents": ["ICO", "IEO", "STO", "tokenomics", "fundraising"],
    "NFT Reports": ["NFT", "non-fungible", "metaverse", "digital art"],
    "Security Reports": ["security", "audit", "vulnerability", "hack"],
    "DeFi Reports": ["DeFi", "liquidity pool", "yield farming"],
    "Taxation Reports": ["tax", "accounting", "audit", "IRS"],
}

CATEGORY_LABELS = list(CATEGORIES.keys())

def get_file_hash(pdf_path):
    try:
        with open(pdf_path, "rb") as f:
            return hashlib.md5(f.read()).hexdigest()
    except Exception as e:
        print(f"Error hashing file {pdf_path}: {e}")
        return None

def check_duplicate_pdfs(folder1, folder2):
    folder2_hashes = {}
    
    if os.path.exists(folder2):
        for file_name in os.listdir(folder2):
            if file_name.endswith(".pdf"):
                file_path = os.path.join(folder2, file_name)
                file_hash = get_file_hash(file_path)
                if file_hash:
                    folder2_hashes[file_hash] = file_path
    
    if os.path.exists(folder1):
        for file_name in os.listdir(folder1):
            if file_name.endswith(".pdf"):
                file_path = os.path.join(folder1, file_name)
                file_hash = get_file_hash(file_path)
                if file_hash in folder2_hashes:
                    print(f"Deleting duplicate file from folder1: {file_path}")
                    os.remove(file_path)
def move_pdfs(folder1, folder2):
    if not os.path.exists(folder2):
        os.makedirs(folder2)
    
    for file_name in os.listdir(folder1):
        if file_name.endswith(".pdf"):
            src_path = os.path.join(folder1, file_name)
            dest_path = os.path.join(folder2, file_name)
            shutil.move(src_path, dest_path)
            print(f"Moved {file_name} to {folder2}")

def extract_text_from_pdf(pdf_path):
    try:
        text = extract_text(pdf_path)
        return text.strip()
    except Exception as e:
        print(f"Error extracting text from {pdf_path}: {e}")
        return ""

def train_model():
    data = []
    labels = []
    
    for category, keywords in CATEGORIES.items():
        for keyword in keywords:
            data.append(keyword)
            labels.append(category)
    
    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(data)
    model = LogisticRegression()
    model.fit(X, labels)
    
    joblib.dump((vectorizer, model), "text_classifier.pkl")
    print("Model trained and saved.")

def load_model():
    return joblib.load("text_classifier.pkl")

def classify_text(text):
    if not text:
        return "Unknown"
    
    vectorizer, model = load_model()
    X = vectorizer.transform([text])
    return model.predict(X)[0]

def classify_pdfs_in_folder(folder_path):
    data = []
    
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".pdf"):
            pdf_path = os.path.join(folder_path, file_name)
            content = extract_text_from_pdf(pdf_path)
            category = classify_text(content)
            data.append([file_name, content, category])
    
    df = pd.DataFrame(data, columns=["File Name", "Content", "Category"])
    return df


folder1 = "small test" 
folder2 = "sample test"
check_duplicate_pdfs(folder1, folder2)
    
if not os.path.exists("text_classifier.pkl"):
    train_model()
    
df=classify_pdfs_in_folder(folder1)
```
