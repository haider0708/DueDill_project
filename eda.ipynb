WEF_Digital_Assets_Regulation_2024.pdf:

This document provides a comprehensive analysis of regulatory approaches, outcomes, and unintended consequences in various jurisdictions, including the European Union, Gibraltar, Hong Kong, Japan, Singapore, Switzerland, the United Arab Emirates, the United Kingdom, and the United States. It examines how different regions regulate financial and technological innovations, particularly in areas such as anti-money laundering (AML) and know-your-customer (KYC) compliance, regulatory and technical sandboxes, decentralized finance (DeFi), and privacy and security policies. The report concludes with a set of recommendations aimed at enhancing regulatory frameworks while balancing innovation and risk management.


```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text() if page.extract_text() else ""  # Extract text
    return text

pdf_text = extract_text_from_pdf("WEF_Digital_Assets_Regulation_2024.pdf")
print(pdf_text)  
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")

def preprocess_text(text):
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    tokens = word_tokenize(text.lower())  # Tokenization & lowercasing
    stop_words = set(stopwords.words("english"))
    tokens = [word for word in tokens if word not in stop_words]  # Remove stopwords
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]  # Lemmatization
    return " ".join(tokens)

cleaned_text = preprocess_text(pdf_text)
```

```python
print("Total Words:", len(cleaned_text.split()))
print("Unique Words:", len(set(cleaned_text.split())))
```

```python
from collections import Counter
import matplotlib.pyplot as plt

word_counts = Counter(cleaned_text.split())
common_words = word_counts.most_common(20)  # Top 20 words

# Plot the top words
words, counts = zip(*common_words)
plt.figure(figsize=(12,5))
plt.bar(words, counts)
plt.xticks(rotation=45)
plt.title("Top 20 Most Common Words")
plt.show()
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color="white").generate(cleaned_text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
