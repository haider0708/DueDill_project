DATA exploration:

```python
import os
import pandas as pd

folder_path = "data"
files = [f for f in os.listdir(folder_path) if f.endswith(".csv")]

data_dict = {}
for file in files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path, parse_dates=["Date"], index_col="Date")
    data_dict[file] = df  # Store each dataframe with its filename as the key

print(f"Loaded {len(data_dict)} CSV files.")
```

Time plot:

```python
import matplotlib.pyplot as plt

for name, df in data_dict.items():
    print(f"\nEDA for {name}:")
    df["Price"] = df["Price"].replace({',': ''}, regex=True).astype(float)
    # Basic stats
    print(df.info())
    print(df.columns)
    print(df["Price"].describe())
    
    # Plot time series
    plt.figure(figsize=(10, 5))
    plt.plot(df["Price"], label=name)
    plt.legend()
    plt.title(f"Time Series Plot: {name}")
    plt.show()
    
    # Check missing values
    print("Missing values:", df.isnull().sum())

```

On remarque des allures differentes dans les series temporelles.

df["Price"] n'est pas toujours de type numrique

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt

for name, df in data_dict.items():
    print(f"\nChecking stationarity for {name}:")
    
    # Perform ADF test
    result = adfuller(df["Price"])
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")

    # Skip decomposition if there are not enough observations for 2 complete cycles
    if len(df) >= 60:  # 2 complete cycles for a period of 30
        # Decomposition
        decomposition = seasonal_decompose(df["Price"], model="additive", period=30)
        decomposition.plot()
        plt.suptitle(f"Decomposition of {name}")
        plt.show()
    else:
        print(f"Skipping decomposition for {name} due to insufficient data.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for name, df in data_dict.items():
    print(f"\nEDA for {name}:")
    if pd.api.types.is_datetime64_any_dtype(df.index):  # Ensure the index is a datetime type
        df['Month'] = df.index.month  # Extract the month from the DateTimeIndex
        
        # Plot the seasonal data (monthly seasonal plot)
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='Month', y='Price', data=df)
    plt.title(f"Seasonal Plot: {name} - Price by Month")
    plt.xlabel("Month")
    plt.ylabel("Price")
    plt.show()
```
